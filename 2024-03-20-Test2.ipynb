```python
import pylab
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats 
from scipy.stats import norm
import seaborn as sns
import math
```

```python
n=630
F=np.zeros((3, 2)); func=np.zeros((1, n)); d=np.zeros((1, n)); func2=np.zeros((1, n))
R=np.asarray([0.2, 0.5, 0.8])
Lambda=1.1

plt.figure()
for i in range(0, 3):
    F=(4*R[i])/((1-R[i])**2)
    for j in range(1, n):
        d=100+(j/220)
        func=(1+(F**2)*(math.sin(((2*np.pi)/Lambda)*(d))))**(-1)
        plt.plot(d, func, marker='o')
        #plt.pause(0.1)
        plt.ylim(-4, 4)

plt.title('Resonator test')

        
plt.show()
```
